Se replicar谩 el modelo para entrenarlo con datos prepandemia y con esto asegurar:

### Origen de los precios predichos ("y_pred")

**驴El modelo de predicci贸n se entrena solo con datos pre-pandemia o incluye post-pandemia?**

La respuesta depende de la estrategia de identificaci贸n y del objetivo causal:

- **Modelo entrenado solo con datos pre-pandemia:**  
  - **Ventaja:** Garantiza que $$y_{pred}$$ refleje 煤nicamente las condiciones estructurales y tendencias previas al shock, evitando que los efectos de la pandemia contaminen el valor de referencia.
  - **Interpretaci贸n:** El "price_premium" entonces mide el desv铆o respecto a un contrafactual puro: 驴cu谩nto se habr铆a esperado que valiera el inmueble si la pandemia no hubiera ocurrido?
  - **Recomendaci贸n metodol贸gica:** Esta es la pr谩ctica recomendada en la literatura causal y de evaluaci贸n de shocks ex贸genos[1][2][3].

### Resumen


- **El modelo para $$y_{pred}$$ debe entrenarse solo con datos pre-pandemia** para asegurar que el contrafactual sea v谩lido y el efecto causal est茅 correctamente identificado, siguiendo las mejores pr谩cticas internacionales en an谩lisis DID de precios inmobiliarios[1][2][3].

A continuaci贸n te presento un pipeline en Python para entrenar desde cero un modelo XGBoost que prediga el precio por metro cuadrado (o la variable objetivo definida en tu configuraci贸n) usando 煤nicamente datos pre-pandemia de la CDMX, siguiendo las mejores pr谩cticas de ingenier铆a de datos y replicando la l贸gica de tu c贸digo base.

```python
# --- 1. Filtrado de datos: Solo CDMX y pre-pandemia ---
import pandas as pd
import numpy as np
import yaml
import xgboost as xgb
from sklearn.model_selection import train_test_split


# 1) Carga y formatea fecha
file_path = '../data/properties_predicted.parquet'
df = pd.read_parquet(file_path)
df['valuation_date'] = pd.to_datetime(df['valuation_date'])

# 2) Calcular quarters_since_first_appraisal en df_cdmx_pre
first_date = df['valuation_date'].min()
#df['quarters_since_first_appraisal'] = (
#    (df['valuation_date'] - first_date).dt.days
#    / (30.4 * 3)
#)

# 2) Calcular months_since_first_appraisal en df_cdmx_pre
df['months_since_first_appraisal'] = (
    (df['valuation_date'] - first_date).dt.days
     / 30.4
)

# 3) Fecha de corte
cutoff = pd.to_datetime('2020-03-01')

# 4) Filtrar todo CDMX
df_cdmx = df[df['state_id'].astype(str) == '09'].copy()

# 5) Crear columna 'period' en df_cdmx
df_cdmx['period'] = np.where(df_cdmx['valuation_date'] < cutoff, 'pre', 'post')

# 6) Ahora extraer s贸lo el pre-pandemia
df_cdmx_pre = df_cdmx[df_cdmx['period'] == 'pre'].copy()
```

```python
# --- 2. Cargar la configuraci贸n YAML del modelo ---
with open("config_xgb.yaml", "r") as f:
    config = yaml.safe_load(f)
config_model = config['model']
recaster_mapper = config.get('categorical_recasters', {})

# Determinar la lista de features usadas por el modelo y preparar X e y
feature_names = list(config_model['features'].keys())
X = df_cdmx_pre[feature_names].copy()

# --- 4. Aplicar recodificaci贸n categ贸rica seg煤n config (si se desea) ---
if recaster_mapper:
    X.replace(recaster_mapper, inplace=True)

# --- 3. Selecci贸n de la variable objetivo (y) ---
# Variable dependiente recomendada: log(precio)
y = np.log(df_cdmx_pre['price'])

# (Opcional: remover la normalizaci贸n por 谩rea en X si se desea usar log(price) directamente)
# Ya tenemos 'saleable_area' en X como control, por lo que mantenerla est谩 bien.

# 3) Filtra prepandemia y split
mask_pre = df_cdmx_pre['valuation_date'] < cutoff
X_pre, y_pre = X[mask_pre], y[mask_pre]
X_train, X_val, y_train, y_val = train_test_split(
    X_pre, y_pre, test_size=0.2, random_state=42
)
```

```python
# 4) Convierte las columnas categ贸ricas a dtype 'category'
cat_cols = ['property_type_id', 'urban_proximity_id']
for df_ in (X_train, X_val):
    for c in cat_cols:
        if c in df_.columns:
            df_[c] = df_[c].astype('category')

# 5) Construye los DMatrix habilitando categor铆as
dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=True)
dval   = xgb.DMatrix(X_val,   label=y_val,   enable_categorical=True)
```

```python
# 6) Par谩metros (id茅nticos al XGBRegressor)
params = {
    'objective':   'reg:squarederror',
    'eval_metric': 'mae',
    'booster':     'gbtree',
    'eta':         0.31,
    'max_depth':   10,
    'lambda':      10,
    'subsample':   1,
    'seed':        42,
    'verbosity':   1
}

# 7) Entrenamiento con early stopping
bst = xgb.train(
    params,
    dtrain,
    num_boost_round=1000,
    evals=[(dtrain, 'train'), (dval, 'validation')],
    early_stopping_rounds=25,
    verbose_eval=50
)

print(f" Mejor iteraci贸n: {bst.best_iteration}")
```

### Evaluaci贸n del modelo hed贸nico (predicciones y residuales)


Una vez entrenado el modelo con datos pre-COVID, podemos aplicarlo tanto a los datos pre (para evaluar ajuste in-sample) como a los datos post-COVID (para obtener predicciones contrafactuales de lo que el precio habr铆a sido sin pandemia). Calculamos las predicciones de log-precio y luego las transformamos a precio real para interpretar errores. Tambi茅n computamos el error o residual de predicci贸n para cada observaci贸n:

```python
# 1) Montar X_all con las mismas features y recasters
feature_names   = list(config_model['features'].keys())
X_all           = df_cdmx[feature_names].copy()

# 2a) Aplicar recasters
if recaster_mapper:
    X_all.replace(recaster_mapper, inplace=True)

# 2b) Convertir a dtype 'category' las columnas categ贸ricas
cat_cols = ['property_type_id', 'urban_proximity_id']
for c in cat_cols:
    if c in X_all.columns:
        X_all[c] = X_all[c].astype('category')

# 3) Crear DMatrix y predecir log-precio
dall = xgb.DMatrix(X_all, enable_categorical=True)
log_price_pred = bst.predict(dall)

# 4) A帽adir predicciones al DataFrame y calcular errores
df_cdmx['log_price_pred'] = log_price_pred
df_cdmx['log_price']      = np.log(df_cdmx['price'])
df_cdmx['price_pred']     = np.exp(log_price_pred)

df_cdmx['error_abs'] = (df_cdmx['price'] - df_cdmx['price_pred']).abs()
df_cdmx['error_log'] = df_cdmx['log_price'] - df_cdmx['log_price_pred']

# 5) Mostrar los primeros registros
print(df_cdmx[['price', 'price_pred', 'error_abs', 'error_log']].head(5))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot

#  Suponemos que df_cdmx ya contiene:
#    'valuation_date', 'price', 'log_price_pred', 'price_pred', 
#    'error_abs' y 'error_log' 

# 1) Sesgo (bias) medio en log y MAPE in-sample (pre-pandemia)
bias = df_cdmx.groupby('period')['error_log'].mean()
mape_pre = (
    (df_cdmx.loc[df_cdmx['period']=='pre', 'error_abs'] /
     df_cdmx.loc[df_cdmx['period']=='pre', 'price'])
    .mean()
)

print("Bias medio (error_log) por periodo:")
print(bias)
print(f"\nMAPE in-sample (pre-pandemia): {mape_pre:.4f}")

# 2) Comprobar umbrales
if abs(bias['pre']) > 0.02 or mape_pre > 0.20:
    print("\n锔 Atenci贸n: El bias o el MAPE in-sample exceden rangos razonables.")
    print("    Revisa transformaciones, interacciones, hiperpar谩metros o modelo.")

# 3) Diagn贸sticos gr谩ficos

# 3.1) Residual vs Predicci贸n (log-scale)
plt.figure()
plt.scatter(df_cdmx['log_price_pred'], df_cdmx['error_log'])
plt.xlabel('Predicted log-price')
plt.ylabel('Residual (log)')
plt.title('Residual vs Predicted (log-scale)')
plt.show()

# 3.2) QQ-plot de residuos
plt.figure()
probplot(df_cdmx['error_log'].dropna(), dist='norm', plot=plt)
plt.title('QQ-plot of Residuals')
plt.show()

# 3.3) Histogramas de error_log por periodo
for periodo in ['pre', 'post']:
    plt.figure()
    data = df_cdmx.loc[df_cdmx['period']==periodo, 'error_log']
    plt.hist(data, bins=50)
    plt.xlabel('Residual (log)')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of error_log ({periodo})')
    plt.show()
```

- mean(error_log) en el periodo pre es -0.000287  0: tu modelo hed贸nico est谩 perfectamente calibrado in-sample, sin sesgos sistem谩ticos.
- MAPE in-sample (pre)  4.6 %
- Un error porcentual medio de s贸lo 4.6 % en precios indica un muy buen ajuste para un modelo no-lineal con XGBoost.
- Bias post-pandemia  +0.1103 log-u.

En el periodo post, el sesgo medio sube a +0.1103 en log-precio. En t茅rminos de precios eso equivale a
$e^{0.1103}111.7%$ es decir, el precio observado est谩 un 11.7 % por encima de lo que predecir铆a el modelo entrenado 煤nicamente con datos pre-COVID.

**锔 Esto NO es un fallo del modelo, sino justamente la brecha contrafactual que el DID aprovecha como estimador del efecto de la pandemia.**

```python
from scipy.stats import ttest_ind

pre_res  = df_cdmx.loc[df_cdmx['period']=='pre',  'error_log']
post_res = df_cdmx.loc[df_cdmx['period']=='post', 'error_log']

t_stat, p_val = ttest_ind(post_res, pre_res, equal_var=False)

effect_pct = np.expm1(post_res.mean() - pre_res.mean()) * 100

print(f"t-statistic = {t_stat:.3f}, p-value = {p_val:.8f}")
print(f"Efecto promedio (post-pre) = {effect_pct:.2f} %")
```

### An谩lisis de la prima de precio (residuo hed贸nico)

Como an谩lisis complementario, calculamos la prima de precio definida como la diferencia entre el logaritmo del precio observado y el logaritmo del precio predicho por el modelo hed贸nico entrenado solo con datos pre-pandemia. Esta m茅trica $log(y) - log(y_{pred})$ representa cu谩nto m谩s alto o m谩s bajo se vendi贸 un inmueble en comparaci贸n con su valor esperado (hed贸nico) dado sus caracter铆sticas. Un valor cercano a 0 indica que se vendi贸 aproximadamente a su valor esperado por el modelo; valores negativos indican que el inmueble se vendi贸 por debajo de lo esperado (descuento), y positivos, por encima (sobreprecio). 

Calculamos esta prima de precio para los inmuebles en el periodo post-COVID, y examinamos su evoluci贸n en el tiempo y por grupo (tratado/control):

```python
import pandas as pd
import numpy as np
from pyproj import Transformer

# --- 2. Calcular log-precios y prima de precio ---
df_cdmx['log_price']      = np.log(df_cdmx['price'])
df_cdmx['log_price_pred'] = np.log(df_cdmx['price_pred'])
df_cdmx['price_premium']  = df_cdmx['log_price'] - df_cdmx['log_price_pred']

# --- 3. Proyecci贸n a CRS m茅trico y distancia euclidiana al centro ---
# transformer 43266372
transformer = Transformer.from_crs("EPSG:4326", "EPSG:6372", always_xy=True)
# proyectar
xy = df_cdmx.apply(
    lambda r: transformer.transform(r['longitude'], r['latitude']),
    axis=1, result_type='expand'
)
df_cdmx[['x_m','y_m']] = xy
# centro proyectado
lon0, lat0 = -99.1332, 19.4326
x0, y0 = transformer.transform(lon0, lat0)
# distancia
df_cdmx['dist_centro_m'] = np.hypot(df_cdmx['x_m']-x0, df_cdmx['y_m']-y0)

# --- 4. Definir zonas en km (convertimos metroskm para etiquetas) ---
df_cdmx['dist_centro_km'] = df_cdmx['dist_centro_m'] / 1000
df_cdmx['zona'] = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, 6, 12, np.inf],
    labels=['C茅ntrico','Intermedio','Perif茅rico']
)

# --- 5. Selecci贸n de hex谩gonos m谩s densos (usando hex_id existente) ---
# contar obs por (hex_id, zona)
counts = (
    df_cdmx
    .groupby(['hex_id','zona'], observed=True)
    .size()
    .reset_index(name='n_obs')
)
# top 10 en C茅ntrico y en Perif茅rico
top_centros   = counts.query("zona=='C茅ntrico'")   .nlargest(10, 'n_obs')['hex_id']
top_perifs    = counts.query("zona=='Perif茅rico'").nlargest(10, 'n_obs')['hex_id']

# --- 6. Crear indicador de tratamiento/control ---
df_cdmx['grupo'] = df_cdmx['hex_id'].map(
    lambda h: 1 if h in set(top_centros)
              else (0 if h in set(top_perifs) else np.nan)
)

# --- 7. Pivot: prima promedio post-COVID por trimestre y grupo ---
df_post = df_cdmx[df_cdmx['period']=='post'].copy()
df_post['month'] = df_post['valuation_date'].dt.to_period('M')

premium_by_q = (
    df_post
    .pivot_table(
        values='price_premium',
        index='month',
        columns='grupo',
        aggfunc='mean',
        observed=True
    )
    .rename(columns={0:'control', 1:'tratado'})
)

print(premium_by_q.head(8))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  1) Partimos de df_cdmx ya con:
#    ['valuation_date','price','price_pred','hex_id','dist_centro_km','grupo']
# y filtrado por state_id == '09'

# 1) Crear el 铆ndice de trimestres (pre + post)
#df_cdmx['quarter'] = df_cdmx['valuation_date'].dt.to_period('Q')
df_cdmx['month'] = df_cdmx['valuation_date'].dt.to_period('M')

# 2) Pivot sobre LA COLUMNA 'grupo', no 'treat'
premium_by_q_all = df_cdmx.pivot_table(
    values='price_premium',
    index='month',
    columns='grupo',          #  aqu铆
    aggfunc='mean',
    observed=True
).rename(columns={0:'control', 1:'tratado'})

cutoff_pre  = pd.Period('2020-02', freq='M')  # 煤ltimo mes pre-COVID (febrero 2020)
cutoff_post = pd.Period('2020-03', freq='M')  # primer mes post-COVID (marzo 2020)

# 4) Si quieres centrarlas en cero usando el promedio pre-COVID:
baseline = premium_by_q_all.loc[:cutoff_pre, ['control','tratado']].mean()
premium_rel = premium_by_q_all - baseline

# 5) Gr谩fica final post-COVID  
post_idx = premium_rel.index >= cutoff_post
```

```python
# 5) Gr谩fica final post-COVID  
plt.figure()
plt.plot(
    premium_rel.index[post_idx].astype(str),
    premium_rel['control'][post_idx],
    marker='o',
    label='Control'
)
plt.plot(
    premium_rel.index[post_idx].astype(str),
    premium_rel['tratado'][post_idx],
    marker='o',
    label='Tratado'
)
plt.axhline(
    0,
    linestyle='--',
    label=(
      f'Promedio pre-COVID\n'
      f'(Ctrl={baseline["control"]:.3f}, '
      f'Trat={baseline["tratado"]:.3f})'
    )
)

plt.xlabel('Mes')                    # ahora es mensual
plt.ylabel(' Prima precio (log y  log 欧)')
plt.title(
  'Prima relativa post-COVID: Tratado vs Control\n'
  '(centrada en promedio pre-COVID)'
)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Al centrar la serie con
premium_rel = premium_by_q_all - baseline
le est谩s quitando justo el valor de baseline (un escalar) a cada punto. Por eso, tras esa resta, el nivel normal de pre-COVID en ambas series pasa a ser 0, y lo 煤nico que ves es la l铆nea punteada de axhline(0).


### Construcci贸n del modelo de Diferencia-en-Diferencias (DID) con controles

Ahora implementamos el modelo econom茅trico DID para cuantificar formalmente el efecto causal del COVID-19 en los precios inmobiliarios de CDMX, siguiendo las recomendaciones. La especificaci贸n base es: 

$$log(precio_{it})=\alpha+\beta(T_i x Post_t)+\delta X_{it}+\mu_{it}+\lambda_t+\epsilon_{it}$$


donde:
$T_i$ es una dummy que indica pertenencia al grupo tratado (inmuebles potencialmente m谩s afectados por la pandemia),
$Post_t$ indica el periodo posterior a marzo 2020,
$X_{it}$ son controles hed贸nicos (caracter铆sticas del inmueble),
$\mu_i$ son efectos fijos por unidad (inmueble) o zona (para capturar heterogeneidad espacial no observable),
$\lambda_t$ son efectos fijos de tiempo (por trimestre/a帽o, para capturar tendencias generales de precios),
$\beta$ es el par谩metro de inter茅s que mide el efecto porcentual de la pandemia en el precio del grupo tratado, neto de controles

#### Definici贸n de grupos tratado y control

La elecci贸n del grupo de tratamiento y control es cr铆tica para una identificaci贸n v谩lida
En un choque global como COVID, ninguna propiedad est谩 completamente no tratada, pero podemos aprovechar diferencias en la intensidad del impacto. Por ejemplo, se sospecha que el impacto en CDMX pudo variar por zona geogr谩fica o por tipo de vivienda:

- Una hip贸tesis es que las viviendas en zonas centrales densas (con uso comercial, oficinal) fueron m谩s afectadas que las viviendas en zonas perif茅ricas residenciales, debido a cambios de preferencias y teletrabajo.
- Otra hip贸tesis es que los departamentos (apartamentos) fueron relativamente menos demandados vs. las casas durante la pandemia, por b煤squeda de espacios m谩s amplios y menor densidad.

Para ilustrar, definiremos el grupo tratado como inmuebles tipo departamento y el control como casas (incluyendo casas en condominio). Esto captura en parte la distinci贸n centro/periferia, ya que en CDMX los departamentos predominan en zonas c茅ntricas y las casas en zonas perif茅ricas, aunque no es una separaci贸n geogr谩fica perfecta. Tambi茅n podr铆amos definir tratado por ubicaci贸n (ej. urban_proximity_id centric vs dem谩s, o ciertas alcald铆as centrales vs otras), pero usaremos tipo de propiedad para claridad en el c贸digo. Definimos entonces treat = 1 si property_type_id corresponde a apartamento (por la recodificaci贸n, ese era el valor 4), y treat = 0 si es casa (valores 2 = casa, 3 = casa en condominio):

```python
# Definir treat (tratado) como 1 si es departamento, 0 si es casa
df_cdmx['treat'] = ((df_cdmx['property_type_id'] == 4)   # apartment 
                    ).astype(int)

# Verificar distribuci贸n de treat
print(df_cdmx['treat'].value_counts())
```

### Opci贸n A: Tratamiento por zona geogr谩fica (zona)


```python
# 2) Definir treat = 1 para zona 'C茅ntrico', 0 en el resto
df_cdmx['treat'] = df_cdmx['zona'].map({
    'C茅ntrico'   : 1,
    'Intermedio' : 'NaN',   # opcional: podr铆as excluir intermedios (NaN)
    'Perif茅rico' : 0
})

# 3) Verifica cu谩ntos de cada grupo hay
print(df_cdmx['treat'].value_counts())
```

### Regresi贸n DID con controles hed贸nicos y efectos fijos

Ahora construimos el conjunto de controles $X_{it}$. Incluiremos variables hed贸nicas relevantes como:
Tama帽o (metros cuadrados vendibles, saleable_area), N煤mero de rec谩maras (bedrooms) y ba帽os (full_bathrooms + half_bathrooms),
Antig眉edad (age_in_months o en a帽os), Amenidades como estacionamientos (parking_lots),


Efectos fijos por zona: usaremos dummies por alcald铆a (municipio) para $\mu_i$ a nivel zona, asumiendo que cada propiedad pertenece a una de las 16 alcald铆as de CDMX (municipality_id). Esto controla diferencias de nivel de precio entre zonas

Efectos fijos de tiempo: usaremos dummies por trimestre (valuation_date convertida a periodo trimestral) para $\lambda_t$, capturando la tendencia general de precios en el tiempo (e.g., efectos macroecon贸micos, tasas de inter茅s bajas que impulsaron los precios en 2020-21)

Construimos y ajustamos la regresi贸n DID con statsmodels (OLS):

```python
import pandas as pd
import numpy as np
from pyproj import Transformer

# Sup贸n que df_cdmx ya contiene: price, valuation_date, longitude, latitude, hex_id, municipality_id, etc.

# 1.1 Proyecci贸n a CRS m茅trico (EPSG:6372) para calcular distancia euclidiana
lon_centro, lat_centro = -99.1332, 19.4326
transformer = Transformer.from_crs("EPSG:4326", "EPSG:6372", always_xy=True)

xy = df_cdmx.apply(lambda r: transformer.transform(r['longitude'], r['latitude']), axis=1, result_type='expand')
df_cdmx[['x_m', 'y_m']] = xy
x0, y0 = transformer.transform(lon_centro, lat_centro)
df_cdmx['dist_centro_m'] = np.hypot(df_cdmx['x_m'] - x0, df_cdmx['y_m'] - y0)
df_cdmx['dist_centro_km'] = df_cdmx['dist_centro_m'] / 1000

# 1.2 Asignaci贸n de zonas y variable de tratamiento a nivel hex谩gono
df_cdmx['zona'] = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, 6, 12, np.inf],
    labels=['C茅ntrico', 'Intermedio', 'Perif茅rico']
)
hex_zonas = (
    df_cdmx[['hex_id','zona']]
    .drop_duplicates(subset='hex_id')            # un par (hex_id, zona) por fila
    .groupby('hex_id')['zona']
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    .reset_index()
    .rename(columns={'zona':'zona_hex'})
)
mapping = {'C茅ntrico': 1, 'Intermedio': np.nan, 'Perif茅rico': 0}
hex_zonas['treat_hex'] = hex_zonas['zona_hex'].map(mapping)
df_cdmx = df_cdmx.merge(
    hex_zonas[['hex_id','treat_hex']],
    on='hex_id',
    how='left'
)

# 1.3 Definir variable post-COVID y trimestre
df_cdmx['post']  = (df_cdmx['valuation_date'] >= pd.Timestamp('2020-03-01')).astype(int)
df_cdmx['month'] = df_cdmx['valuation_date'].dt.to_period('M')  # mantenlo como Period si vas a slice

# 1.4 Filtrar solo hex谩gonos c茅ntricos y perif茅ricos
df_did = df_cdmx[df_cdmx['treat'].isin([0, 1])].copy()
```

```python
import statsmodels.formula.api as smf

# 2.1 F贸rmula con controles hed贸nicos y efectos fijos de zona (alcald铆a) y tiempo (trimestre)
df_did['month'] = df_did['month'].astype(str)
formula = (
    "np.log(price) ~ treat * post"
    " + bedrooms + full_bathrooms + half_bathrooms"
    " + saleable_area + age_in_months + parking_lots"
    " + C(municipality_id) + C(month)"
)

# 2.2 Ajuste del modelo con errores est谩ndar agrupados por hex谩gono
did = smf.ols(formula, data=df_did).fit(
    cov_type="cluster",
    cov_kwds={"groups": df_did["hex_id"]}
)
print(did.summary())
```

Este efecto es estad铆sticamente significativo al 5%.
