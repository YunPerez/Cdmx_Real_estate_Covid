# Replica de Modelo XGBoost

Se replicará el modelo para entrenarlo con datos prepandemia y con esto asegurar:

### Origen de los precios predichos ("y_pred")

**¿El modelo de predicción se entrena solo con datos pre-pandemia o incluye post-pandemia?**

La respuesta depende de la estrategia de identificación y del objetivo causal:

- **Modelo entrenado solo con datos pre-pandemia:**  
  - **Ventaja:** Garantiza que $$y_{pred}$$ refleje únicamente las condiciones estructurales y tendencias previas al shock, evitando que los efectos de la pandemia contaminen el valor de referencia.
  - **Interpretación:** El "price_premium" entonces mide el desvío respecto a un contrafactual puro: ¿cuánto se habría esperado que valiera el inmueble si la pandemia no hubiera ocurrido?
  - **Recomendación metodológica:** Esta es la práctica recomendada en la literatura causal y de evaluación de shocks exógenos[1][2][3].

### Resumen


- **El modelo para $$y_{pred}$$ debe entrenarse solo con datos pre-pandemia** para asegurar que el contrafactual sea válido y el efecto causal esté correctamente identificado, siguiendo las mejores prácticas internacionales en análisis DID de precios inmobiliarios[1][2][3].

A continuación te presento un pipeline en Python para entrenar desde cero un modelo XGBoost que prediga el precio por metro cuadrado (o la variable objetivo definida en tu configuración) usando únicamente datos pre-pandemia de la CDMX, siguiendo las mejores prácticas de ingeniería de datos y replicando la lógica de tu código base.

```python
# --- 1. Filtrado de datos: Solo CDMX y pre-pandemia ---
import pandas as pd
import numpy as np
import yaml
import xgboost as xgb
from sklearn.model_selection import train_test_split


# 1) Carga y formatea fecha
file_path = '../data/properties_predicted.parquet'
df = pd.read_parquet(file_path)
df['valuation_date'] = pd.to_datetime(df['valuation_date'])

# 2) Calcular quarters_since_first_appraisal en df_cdmx_pre
first_date = df['valuation_date'].min()
#df['quarters_since_first_appraisal'] = (
#    (df['valuation_date'] - first_date).dt.days
#    / (30.4 * 3)
#)

# 2) Calcular months_since_first_appraisal en df_cdmx_pre
df['months_since_first_appraisal'] = (
    (df['valuation_date'] - first_date).dt.days
     / 30.4
)

# 3) Fecha de corte
cutoff = pd.to_datetime('2020-03-01')

# 4) Filtrar todo CDMX
df_cdmx = df[df['state_id'].astype(str) == '09'].copy()

# 5) Crear columna 'period' en df_cdmx
df_cdmx['period'] = np.where(df_cdmx['valuation_date'] < cutoff, 'pre', 'post')

# 6) Ahora extraer sólo el pre-pandemia
df_cdmx_pre = df_cdmx[df_cdmx['period'] == 'pre'].copy()
```

```python
# --- 2. Cargar la configuración YAML del modelo ---
with open("config_xgb.yaml", "r") as f:
    config = yaml.safe_load(f)
config_model = config['model']
recaster_mapper = config.get('categorical_recasters', {})

# Determinar la lista de features usadas por el modelo y preparar X e y
feature_names = list(config_model['features'].keys())
X = df_cdmx_pre[feature_names].copy()

# --- 4. Aplicar recodificación categórica según config (si se desea) ---
if recaster_mapper:
    X.replace(recaster_mapper, inplace=True)

# --- 3. Selección de la variable objetivo (y) ---
# Variable dependiente recomendada: log(precio)
y = np.log(df_cdmx_pre['price'])

# (Opcional: remover la normalización por área en X si se desea usar log(price) directamente)
# Ya tenemos 'saleable_area' en X como control, por lo que mantenerla está bien.

# 3) Filtra pre‐pandemia y split
mask_pre = df_cdmx_pre['valuation_date'] < cutoff
X_pre, y_pre = X[mask_pre], y[mask_pre]
X_train, X_val, y_train, y_val = train_test_split(
    X_pre, y_pre, test_size=0.2, random_state=42
)
```

```python
# 4) Convierte las columnas categóricas a dtype 'category'
cat_cols = ['property_type_id', 'urban_proximity_id']
for df_ in (X_train, X_val):
    for c in cat_cols:
        if c in df_.columns:
            df_[c] = df_[c].astype('category')

# 5) Construye los DMatrix habilitando categorías
dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=True)
dval   = xgb.DMatrix(X_val,   label=y_val,   enable_categorical=True)
```

```python
# 6) Parámetros (idénticos al XGBRegressor)
params = {
    'objective':   'reg:squarederror',
    'eval_metric': 'mae',
    'booster':     'gbtree',
    'eta':         0.31,
    'max_depth':   10,
    'lambda':      10,
    'subsample':   1,
    'seed':        42,
    'verbosity':   1
}

# 7) Entrenamiento con early stopping
bst = xgb.train(
    params,
    dtrain,
    num_boost_round=1000,
    evals=[(dtrain, 'train'), (dval, 'validation')],
    early_stopping_rounds=25,
    verbose_eval=50
)

print(f"🏆 Mejor iteración: {bst.best_iteration}")
```

### Evaluación del modelo hedónico (predicciones y residuales)


Una vez entrenado el modelo con datos pre-COVID, podemos aplicarlo tanto a los datos pre (para evaluar ajuste in-sample) como a los datos post-COVID (para obtener predicciones contrafactuales de lo que el precio habría sido sin pandemia). Calculamos las predicciones de log-precio y luego las transformamos a precio real para interpretar errores. También computamos el error o residual de predicción para cada observación:

```python
# 1) Montar X_all con las mismas features y recasters
feature_names   = list(config_model['features'].keys())
X_all           = df_cdmx[feature_names].copy()

# 2a) Aplicar recasters
if recaster_mapper:
    X_all.replace(recaster_mapper, inplace=True)

# 2b) Convertir a dtype 'category' las columnas categóricas
cat_cols = ['property_type_id', 'urban_proximity_id']
for c in cat_cols:
    if c in X_all.columns:
        X_all[c] = X_all[c].astype('category')

# 3) Crear DMatrix y predecir log-precio
dall = xgb.DMatrix(X_all, enable_categorical=True)
log_price_pred = bst.predict(dall)

# 4) Añadir predicciones al DataFrame y calcular errores
df_cdmx['log_price_pred'] = log_price_pred
df_cdmx['log_price']      = np.log(df_cdmx['price'])
df_cdmx['price_pred']     = np.exp(log_price_pred)

df_cdmx['error_abs'] = (df_cdmx['price'] - df_cdmx['price_pred']).abs()
df_cdmx['error_log'] = df_cdmx['log_price'] - df_cdmx['log_price_pred']

# 5) Mostrar los primeros registros
print(df_cdmx[['price', 'price_pred', 'error_abs', 'error_log']].head(5))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot

# — Suponemos que df_cdmx ya contiene:
#    'valuation_date', 'price', 'log_price_pred', 'price_pred', 
#    'error_abs' y 'error_log' —

# 1) Sesgo (bias) medio en log y MAPE in-sample (pre-pandemia)
bias = df_cdmx.groupby('period')['error_log'].mean()
mape_pre = (
    (df_cdmx.loc[df_cdmx['period']=='pre', 'error_abs'] /
     df_cdmx.loc[df_cdmx['period']=='pre', 'price'])
    .mean()
)

print("Bias medio (error_log) por periodo:")
print(bias)
print(f"\nMAPE in-sample (pre-pandemia): {mape_pre:.4f}")

# 2) Comprobar umbrales
if abs(bias['pre']) > 0.02 or mape_pre > 0.20:
    print("\n⚠️ Atención: El bias o el MAPE in-sample exceden rangos razonables.")
    print("   → Revisa transformaciones, interacciones, hiperparámetros o modelo.")

# 3) Diagnósticos gráficos

# 3.1) Residual vs Predicción (log-scale)
plt.figure()
plt.scatter(df_cdmx['log_price_pred'], df_cdmx['error_log'])
plt.xlabel('Predicted log-price')
plt.ylabel('Residual (log)')
plt.title('Residual vs Predicted (log-scale)')
plt.show()

# 3.2) QQ-plot de residuos
plt.figure()
probplot(df_cdmx['error_log'].dropna(), dist='norm', plot=plt)
plt.title('QQ-plot of Residuals')
plt.show()

# 3.3) Histogramas de error_log por periodo
for periodo in ['pre', 'post']:
    plt.figure()
    data = df_cdmx.loc[df_cdmx['period']==periodo, 'error_log']
    plt.hist(data, bins=50)
    plt.xlabel('Residual (log)')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of error_log ({periodo})')
    plt.show()
```

- Bias pre-pandemia ≈ 0
- mean(error_log) en el periodo pre es -0.000287 ≈ 0: tu modelo hedónico está perfectamente calibrado in-sample, sin sesgos sistemáticos.
- MAPE in-sample (pre) ≈ 4.6 %
- Un error porcentual medio de sólo 4.6 % en precios indica un muy buen ajuste para un modelo no-lineal con XGBoost.
- Bias post-pandemia ≈ +0.1103 log-u.

En el periodo post, el sesgo medio sube a +0.1103 en log-precio. En términos de precios eso equivale a
$e^{0.1103}−1≈11.7%$ es decir, el precio observado está un 11.7 % por encima de lo que predeciría el modelo entrenado únicamente con datos pre-COVID.

**⚠️ Esto NO es un “fallo” del modelo, sino justamente la brecha contrafactual que el DID aprovecha como estimador del efecto de la pandemia.**

```python
from scipy.stats import ttest_ind

pre_res  = df_cdmx.loc[df_cdmx['period']=='pre',  'error_log']
post_res = df_cdmx.loc[df_cdmx['period']=='post', 'error_log']

t_stat, p_val = ttest_ind(post_res, pre_res, equal_var=False)

effect_pct = np.expm1(post_res.mean() - pre_res.mean()) * 100

print(f"t-statistic = {t_stat:.3f}, p-value = {p_val:.8f}")
print(f"Efecto promedio (post-pre) = {effect_pct:.2f} %")
```

### Análisis de la “prima de precio” (residuo hedónico)

Como análisis complementario, calculamos la “prima de precio” definida como la diferencia entre el logaritmo del precio observado y el logaritmo del precio predicho por el modelo hedónico entrenado solo con datos pre-pandemia. Esta métrica $log(y) - log(y_{pred})$ representa cuánto más alto o más bajo se vendió un inmueble en comparación con su valor esperado (hedónico) dado sus características. Un valor cercano a 0 indica que se vendió aproximadamente a su valor “esperado” por el modelo; valores negativos indican que el inmueble se vendió por debajo de lo esperado (descuento), y positivos, por encima (sobreprecio). 

Calculamos esta prima de precio para los inmuebles en el periodo post-COVID, y examinamos su evolución en el tiempo y por grupo (tratado/control):

```python
import pandas as pd
import numpy as np
from pyproj import Transformer

# --- 2. Calcular log-precios y prima de precio ---
df_cdmx['log_price']      = np.log(df_cdmx['price'])
df_cdmx['log_price_pred'] = np.log(df_cdmx['price_pred'])
df_cdmx['price_premium']  = df_cdmx['log_price'] - df_cdmx['log_price_pred']

# --- 3. Proyección a CRS métrico y distancia euclidiana al centro ---
# transformer 4326→6372
transformer = Transformer.from_crs("EPSG:4326", "EPSG:6372", always_xy=True)
# proyectar
xy = df_cdmx.apply(
    lambda r: transformer.transform(r['longitude'], r['latitude']),
    axis=1, result_type='expand'
)
df_cdmx[['x_m','y_m']] = xy
# centro proyectado
lon0, lat0 = -99.1332, 19.4326
x0, y0 = transformer.transform(lon0, lat0)
# distancia
df_cdmx['dist_centro_m'] = np.hypot(df_cdmx['x_m']-x0, df_cdmx['y_m']-y0)

# --- 4. Definir zonas en km (convertimos metros→km para etiquetas) ---
df_cdmx['dist_centro_km'] = df_cdmx['dist_centro_m'] / 1000
df_cdmx['zona'] = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, 6, 12, np.inf],
    labels=['Céntrico','Intermedio','Periférico']
)

# --- 5. Selección de hexágonos más densos (usando hex_id existente) ---
# contar obs por (hex_id, zona)
counts = (
    df_cdmx
    .groupby(['hex_id','zona'], observed=True)
    .size()
    .reset_index(name='n_obs')
)
# top 10 en Céntrico y en Periférico
top_centros   = counts.query("zona=='Céntrico'")   .nlargest(10, 'n_obs')['hex_id']
top_perifs    = counts.query("zona=='Periférico'").nlargest(10, 'n_obs')['hex_id']

# --- 6. Crear indicador de tratamiento/control ---
df_cdmx['grupo'] = df_cdmx['hex_id'].map(
    lambda h: 1 if h in set(top_centros)
              else (0 if h in set(top_perifs) else np.nan)
)

# --- 7. Pivot: prima promedio post-COVID por trimestre y grupo ---
df_post = df_cdmx[df_cdmx['period']=='post'].copy()
df_post['month'] = df_post['valuation_date'].dt.to_period('M')

premium_by_q = (
    df_post
    .pivot_table(
        values='price_premium',
        index='month',
        columns='grupo',
        aggfunc='mean',
        observed=True
    )
    .rename(columns={0:'control', 1:'tratado'})
)

print(premium_by_q.head(8))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# — 1) Partimos de df_cdmx ya con:
#    ['valuation_date','price','price_pred','hex_id','dist_centro_km','grupo']
# y filtrado por state_id == '09'

# 1) Crear el índice de trimestres (pre + post)
#df_cdmx['quarter'] = df_cdmx['valuation_date'].dt.to_period('Q')
df_cdmx['month'] = df_cdmx['valuation_date'].dt.to_period('M')

# 2) Pivot sobre LA COLUMNA 'grupo', no 'treat'
premium_by_q_all = df_cdmx.pivot_table(
    values='price_premium',
    index='month',
    columns='grupo',          # ← aquí
    aggfunc='mean',
    observed=True
).rename(columns={0:'control', 1:'tratado'})

cutoff_pre  = pd.Period('2020-02', freq='M')  # último mes pre-COVID (febrero 2020)
cutoff_post = pd.Period('2020-03', freq='M')  # primer mes post-COVID (marzo 2020)

# 4) Si quieres centrarlas en cero usando el promedio pre-COVID:
baseline = premium_by_q_all.loc[:cutoff_pre, ['control','tratado']].mean()
premium_rel = premium_by_q_all - baseline

# 5) Gráfica final post-COVID  
post_idx = premium_rel.index >= cutoff_post
```

```python
# 5) Gráfica final post-COVID  
plt.figure()
plt.plot(
    premium_rel.index[post_idx].astype(str),
    premium_rel['control'][post_idx],
    marker='o',
    label='Control'
)
plt.plot(
    premium_rel.index[post_idx].astype(str),
    premium_rel['tratado'][post_idx],
    marker='o',
    label='Tratado'
)
plt.axhline(
    0,
    linestyle='--',
    label=(
      f'Promedio pre-COVID\n'
      f'(Ctrl={baseline["control"]:.3f}, '
      f'Trat={baseline["tratado"]:.3f})'
    )
)

plt.xlabel('Mes')                    # ahora es mensual
plt.ylabel('Δ Prima precio (log y – log ŷ)')
plt.title(
  'Prima relativa post-COVID: Tratado vs Control\n'
  '(centrada en promedio pre-COVID)'
)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

premium_rel = premium_by_q_all - baseline
le estás quitando justo el valor de baseline (un escalar) a cada punto. Por eso, tras esa resta, el “nivel normal” de pre-COVID en ambas series pasa a ser 0, y lo único que ves es la línea punteada de axhline(0).


### Construcción del modelo de Diferencia-en-Diferencias (DID) con controles

Ahora implementamos el modelo econométrico DID para cuantificar formalmente el efecto causal del COVID-19 en los precios inmobiliarios de CDMX, siguiendo las recomendaciones. La especificación base es: 

$$log(precio_{it})=\alpha+\beta(T_i x Post_t)+\delta X_{it}+\mu_{it}+\lambda_t+\epsilon_{it}$$


donde:
$T_i$ es una dummy que indica pertenencia al grupo tratado (inmuebles potencialmente más afectados por la pandemia),
$Post_t$ indica el periodo posterior a marzo 2020,
$X_{it}$ son controles hedónicos (características del inmueble),
$\mu_i$ son efectos fijos por unidad (inmueble) o zona (para capturar heterogeneidad espacial no observable),
$\lambda_t$ son efectos fijos de tiempo (por trimestre/año, para capturar tendencias generales de precios),
$\beta$ es el parámetro de interés que mide el efecto porcentual de la pandemia en el precio del grupo tratado, neto de controles

#### Definición de grupos tratado y control

La elección del grupo de tratamiento y control es crítica para una identificación válida
En un choque global como COVID, ninguna propiedad está completamente “no tratada”, pero podemos aprovechar diferencias en la intensidad del impacto. Por ejemplo, se sospecha que el impacto en CDMX pudo variar por zona geográfica o por tipo de vivienda:

- Una hipótesis es que las viviendas en zonas centrales densas (con uso comercial, oficinal) fueron más afectadas que las viviendas en zonas periféricas residenciales, debido a cambios de preferencias y teletrabajo.
- Otra hipótesis es que los departamentos (apartamentos) fueron relativamente menos demandados vs. las casas durante la pandemia, por búsqueda de espacios más amplios y menor densidad.

Para ilustrar, definiremos el grupo tratado como inmuebles tipo departamento y el control como casas (incluyendo casas en condominio). Esto captura en parte la distinción centro/periferia, ya que en CDMX los departamentos predominan en zonas céntricas y las casas en zonas periféricas, aunque no es una separación geográfica perfecta. También podríamos definir tratado por ubicación (ej. urban_proximity_id centric vs demás, o ciertas alcaldías centrales vs otras), pero usaremos tipo de propiedad para claridad en el código. Definimos entonces treat = 1 si property_type_id corresponde a apartamento (por la recodificación, ese era el valor 4), y treat = 0 si es casa (valores 2 = casa, 3 = casa en condominio):

```python
# Definir treat (tratado) como 1 si es departamento, 0 si es casa
df_cdmx['treat'] = ((df_cdmx['property_type_id'] == 4)   # apartment 
                    ).astype(int)

# Verificar distribución de treat
print(df_cdmx['treat'].value_counts())
```

### Opción A: Tratamiento por zona geográfica (zona)


```python
# 2) Definir treat = 1 para zona 'Céntrico', 0 en el resto
df_cdmx['treat'] = df_cdmx['zona'].map({
    'Céntrico'   : 1,
    'Intermedio' : 'NaN',   # opcional: podrías excluir intermedios (NaN)
    'Periférico' : 0
})

# 3) Verifica cuántos de cada grupo hay
print(df_cdmx['treat'].value_counts())
```

### Opción B (avanzada): Tratamiento por alcaldías centrales

```python
# definimos la lista de IDs centrales (ej.: '001'='Álvaro Obregón', '003'='Benito Juárez', etc.)
central_ids = ['001','003','004','006','011']

# treat = 1 si municipality_id está en central_ids, 0 en otro caso
df_cdmx['treat'] = df_cdmx['municipality_id'].astype(str).isin(central_ids).astype(int)

# validamos
print(df_cdmx['treat'].value_counts())
```

### Regresión DID con controles hedónicos y efectos fijos

Ahora construimos el conjunto de controles $X_{it}$. Incluiremos variables hedónicas relevantes como:
Tamaño (metros cuadrados vendibles, saleable_area), Número de recámaras (bedrooms) y baños (full_bathrooms + half_bathrooms),
Antigüedad (age_in_months o en años), Amenidades como estacionamientos (parking_lots),


Efectos fijos por zona: usaremos dummies por alcaldía (municipio) para $\mu_i$ a nivel zona, asumiendo que cada propiedad pertenece a una de las 16 alcaldías de CDMX (municipality_id). Esto controla diferencias de nivel de precio entre zonas

Efectos fijos de tiempo: usaremos dummies por trimestre (valuation_date convertida a periodo trimestral) para $\lambda_t$, capturando la tendencia general de precios en el tiempo (e.g., efectos macroeconómicos, tasas de interés bajas que impulsaron los precios en 2020-21)

Construimos y ajustamos la regresión DID con statsmodels (OLS):

```python
import pandas as pd
import numpy as np
from pyproj import Transformer

# Supón que df_cdmx ya contiene: price, valuation_date, longitude, latitude, hex_id, municipality_id, etc.

# 1.1 Proyección a CRS métrico (EPSG:6372) para calcular distancia euclidiana
lon_centro, lat_centro = -99.1332, 19.4326
transformer = Transformer.from_crs("EPSG:4326", "EPSG:6372", always_xy=True)

xy = df_cdmx.apply(lambda r: transformer.transform(r['longitude'], r['latitude']), axis=1, result_type='expand')
df_cdmx[['x_m', 'y_m']] = xy
x0, y0 = transformer.transform(lon_centro, lat_centro)
df_cdmx['dist_centro_m'] = np.hypot(df_cdmx['x_m'] - x0, df_cdmx['y_m'] - y0)
df_cdmx['dist_centro_km'] = df_cdmx['dist_centro_m'] / 1000

# 1.2 Asignación de zonas y variable de tratamiento a nivel hexágono
df_cdmx['zona'] = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, 6, 12, np.inf],
    labels=['Céntrico', 'Intermedio', 'Periférico']
)
hex_zonas = (
    df_cdmx[['hex_id','zona']]
    .drop_duplicates(subset='hex_id')            # un par (hex_id, zona) por fila
    .groupby('hex_id')['zona']
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    .reset_index()
    .rename(columns={'zona':'zona_hex'})
)
mapping = {'Céntrico': 1, 'Intermedio': np.nan, 'Periférico': 0}
hex_zonas['treat_hex'] = hex_zonas['zona_hex'].map(mapping)
df_cdmx = df_cdmx.merge(
    hex_zonas[['hex_id','treat_hex']],
    on='hex_id',
    how='left'
)

# 1.3 Definir variable post-COVID y trimestre
df_cdmx['post']  = (df_cdmx['valuation_date'] >= pd.Timestamp('2020-03-01')).astype(int)
df_cdmx['month'] = df_cdmx['valuation_date'].dt.to_period('M')  # mantenlo como Period si vas a slice

# 1.4 Filtrar solo hexágonos céntricos y periféricos
df_did = df_cdmx[df_cdmx['treat'].isin([0, 1])].copy()
```

```python
import statsmodels.formula.api as smf

# 2.1 Fórmula con controles hedónicos y efectos fijos de zona (alcaldía) y tiempo (trimestre)
df_did['month'] = df_did['month'].astype(str)
formula = (
    "np.log(price) ~ treat * post"
    " + bedrooms + full_bathrooms + half_bathrooms"
    " + saleable_area + age_in_months + parking_lots"
    " + C(municipality_id) + C(month)"
)

# 2.2 Ajuste del modelo con errores estándar agrupados por hexágono
did = smf.ols(formula, data=df_did).fit(
    cov_type="cluster",
    cov_kwds={"groups": df_did["hex_id"]}
)
print(did.summary())
```

## Interpretación de la regresión DID con controles hedónicos y efectos fijos

### Resumen del modelo

- **Variable dependiente:** Logaritmo natural del precio de venta (`np.log(price)`).
- **R² ajustado:** 0.804 (el modelo explica el 80.4% de la varianza de los precios, lo cual es muy alto y esperable al incluir efectos fijos y controles detallados).
- **Controles incluidos:** Variables hedónicas (recámaras, baños, área vendible, antigüedad, estacionamientos), efectos fijos por alcaldía (`C(municipality_id)`) y por trimestre (`C(quarter)`).
- **Variables clave DID:**  
  - `treat`: 1 para hexágonos céntricos, 0 para periféricos.
  - `post`: 1 para periodo post-COVID (después de marzo 2020), 0 para pre-COVID.
  - `treat:post`: interacción DID que estima el efecto causal.

### Resultados principales

#### Efecto causal DID (`treat:post`)

- **Coeficiente:** 0.0450
- **Error estándar:** 0.019
- **z:** 2.41
- **p-valor:** 0.016
- **Intervalo 95%:** [0.008, 0.082]

**Interpretación:**  
El coeficiente de la interacción indica que, tras el inicio de la pandemia, los hexágonos céntricos experimentaron un incremento adicional del 4.5% en el precio promedio de los inmuebles respecto a los hexágonos periféricos, controlando por características del inmueble y tendencias específicas de zona y tiempo. Este efecto es estadísticamente significativo al 5%.

#### Otras variables relevantes

- **Controles hedónicos:**  
  - Más baños completos y medios, mayor área vendible y más estacionamientos están asociados a precios significativamente más altos.
  - La antigüedad en meses tiene un efecto negativo pequeño pero significativo.
- **Efectos fijos de alcaldía y trimestre:**  
  - La mayoría de las dummies de alcaldía y trimestre son significativas, reflejando diferencias estructurales de precio entre zonas y momentos del tiempo.

### Diagnóstico del modelo

- **Errores estándar robustos:** Se usaron errores agrupados por hexágono, lo que es adecuado para datos espaciales y paneles repetidos.
- **Condición numérica:** El número de condición es alto (7,170), lo que sugiere cierta multicolinealidad, pero no hay evidencia de problemas graves (los coeficientes y errores estándar están bien definidos y la matriz no es singular).
- **Normalidad de residuos:** El test de Jarque-Bera y la curtosis indican que los residuos no son perfectamente normales, pero esto es común en grandes muestras inmobiliarias.

### Conclusión y relevancia causal

- **El efecto DID estimado es robusto y estadísticamente significativo:** El COVID-19 tuvo un impacto positivo y diferencial en los precios de zonas céntricas respecto a las periféricas, incluso controlando por características del inmueble y efectos fijos de zona y tiempo.
- **El modelo es adecuado para inferencia causal:** La especificación sigue las mejores prácticas de la literatura econométrica y urbana, usando controles hedónicos, efectos fijos y errores estándar robustos.
- **Magnitud:** Un 4.5% de incremento relativo en precios es relevante en el contexto inmobiliario urbano y consistente con hallazgos recientes en mercados afectados por shocks exógenos.
