## Creación de series de tiempo de premio de precio ajustado por calidad a nivel hexágono y colonia

```python
import pandas as pd
import numpy as np
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# --- 1. Cargar datos y preparar columnas básicas ---
file_path = '../data/properties_predicted.parquet'
df = pd.read_parquet(file_path)

# Asegura formato de fecha
df['valuation_date'] = pd.to_datetime(df['valuation_date'])

# --- 2. Calcular premio de precio (price premium) ---
df['log_y'] = np.log(df['y'])
df['log_y_pred'] = np.log(df['y_pred'])
df['price_premium'] = df['log_y'] - df['log_y_pred']

# --- 3. Crear columna de periodo mensual ---
df['period'] = df['valuation_date'].dt.to_period('M')
df_cdmx = df[df['state_id'] == "09"].copy()

# --- 4. Asignar zona geoespacial y grupo de tratamiento/control ---
# -------------------------------------------------------------------

# --- a) Definir centro urbano y calcular distancia ---
centro_cdmx = {'lon': -99.1332, 'lat': 19.4326}

def haversine_np(lon1, lat1, lon2, lat2):
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

df_cdmx['dist_centro_km'] = haversine_np(
    df_cdmx['longitude'], df_cdmx['latitude'],
    centro_cdmx['lon'], centro_cdmx['lat']
)

# --- b) Asignar zona y grupo ---
umbral_centrico = 6
umbral_intermedio = 12

df_cdmx['zona'] = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, umbral_centrico, umbral_intermedio, np.inf],
    labels=['Céntrico', 'Intermedio', 'Periférico']
)

# --- Grupo: 1 = tratamiento (Céntrico), 0 = control (Periférico), Intermedio opcional ---
df_cdmx['grupo'] = df_cdmx['zona'].map({'Céntrico': 1, 'Periférico': 0})
```

```python
# --- 5. Series de tiempo por hexágono (H3) y colonia ---
series_hex = (
    df_cdmx.groupby(['hex_id', 'period'])
    .agg(
        avg_price_premium=('price_premium', 'mean'),
        volume=('y', 'count')
    )
    .reset_index()
)
series_hex['period'] = series_hex['period'].dt.to_timestamp()
print("Series por hexágono:")
print(series_hex.head())

series_colonia = (
    df_cdmx.groupby(['neighborhood', 'period'])
    .agg(
        avg_price_premium=('price_premium', 'mean'),
        volume=('y', 'count')
    )
    .reset_index()
)
series_colonia['period'] = series_colonia['period'].dt.to_timestamp()
print("Series por colonia:")
print(series_colonia.head())
```

```python
# --- 6. Homogeneización y filtrado de calidad ---
def filtra_panel(df, id_col, min_pre=6, min_post=6, corte='2020-03-01'):
    pre = df[df['period'] < corte].groupby(id_col)['volume'].sum()
    post = df[df['period'] >= corte].groupby(id_col)['volume'].sum()
    validos = pre[pre >= min_pre].index.intersection(post[post >= min_post].index)
    return df[df[id_col].isin(validos)].copy()

series_hex_f = filtra_panel(series_hex, 'hex_id')
series_colonia_f = filtra_panel(series_colonia, 'neighborhood')
```

## ¿Cómo tratar la categoría "Intermedio" para un análisis causal robusto?

La forma en que se maneja la zona **Intermedio** es crucial para la validez y la interpretación de los resultados en análisis causal espacial, especialmente en modelos DID, espacio de estados o VAR. La literatura internacional y nacional proporciona recomendaciones claras y respaldadas para decidir cómo tratar esta categoría.

### Opciones principales según la literatura

#### 1. Excluir "Intermedio" del análisis causal principal
- **Recomendado cuando:** Buscas una comparación clara y robusta entre extremos (Céntrico vs. Periférico), como hacen Zeng & Yi (2022) y Tsai et al. (2022).
- **Ventajas:** 
  - Facilita la interpretación causal y la validez del supuesto de tendencias paralelas.
  - Reduce el riesgo de contaminación entre grupos (spillover).
- **Desventajas:** 
  - Pierdes información potencial sobre zonas de transición.
- **Implementación:** 
  - Usa solo Céntrico (tratamiento) y Periférico (control) para el DID principal.
  - Analiza Intermedio en un análisis descriptivo o de robustez por separado.

#### 2. Incluir "Intermedio" como grupo control adicional o grupo intermedio
- **Recomendado cuando:** El objetivo es capturar gradientes de impacto o explorar heterogeneidad espacial, como en Tomal & Helbich (2022).
- **Opciones:**
  - **a) Control extendido:** Une Intermedio y Periférico como grupo control si sus tendencias pre-pandemia son paralelas y similares.
  - **b) Triple diferencia (DDD):** Modela tres grupos (Céntrico, Intermedio, Periférico) y compara los efectos relativos.
  - **c) Análisis multinivel:** Usa modelos de efectos mixtos para estimar diferencias entre los tres grupos y su variabilidad.
- **Ventajas:** 
  - Permite examinar si el efecto causal muestra un gradiente espacial.
  - Detecta patrones de spillover o amortiguamiento del shock.
- **Desventajas:**
  - Mayor complejidad y posibles problemas de interpretación si las tendencias pre no son paralelas entre los tres grupos.

#### 3. Análisis de robustez y heterogeneidad
- **Recomendado por:** La mayoría de los estudios internacionales y nacionales.
- **Estrategia:** 
  - Realiza el análisis principal solo con Céntrico vs. Periférico.
  - Luego, incluye Intermedio para verificar si los resultados son robustos o si hay evidencia de gradiente o segmentación espacial.
  - Reporta explícitamente cualquier diferencia o patrón relevante.

### Recomendación práctica y fundamentada

- **Para el análisis causal principal (DID, espacio de estados):**
  - **Deja Intermedio fuera** y compara solo Céntrico vs. Periférico para maximizar la robustez y la claridad causal[1][2][3].
- **Para análisis de heterogeneidad y robustez:**
  - **Incluye Intermedio** como grupo adicional, ya sea como control extendido o en modelos multinivel, para explorar gradientes y segmentación[2].
- **Siempre valida tendencias paralelas** entre los grupos antes de incluir Intermedio en el análisis causal.

### Respaldo en la literatura

- **Zeng & Yi (2022):** Comparan solo centro vs. periferia para el efecto causal principal y analizan zonas intermedias en robustez y heterogeneidad.
- **Tomal & Helbich (2022):** Usan modelos multinivel y segmentación espacial para explorar gradientes y patrones intermedios, pero reportan el efecto causal principal entre extremos.
- **Tsai et al. (2022):** Recomiendan excluir zonas intermedias del análisis principal si no cumplen tendencias paralelas o presentan heterogeneidad significativa.

**En síntesis:**  
- **Para un análisis causal robusto y comparable internacionalmente, trata Intermedio como grupo de robustez y heterogeneidad, pero no como parte del análisis causal principal entre tratamiento y control.**  
- **Incluye los resultados de Intermedio en anexos, análisis descriptivos o modelos multinivel para enriquecer la discusión y la interpretación del gradiente espacial.**

```python
# --- 7. Asignación consistente de grupo a hexágonos y colonias (excluyendo Intermedio) ---
# --- Para hexágonos ---
mapping_hex = (
    df_cdmx[~df_cdmx['grupo'].isna()]
    .groupby('hex_id')
    .agg(grupo=('grupo', lambda x: x.value_counts().idxmax()))
    .reset_index()
)

# --- Para colonias ---
mapping_colonia = (
    df_cdmx[~df_cdmx['grupo'].isna()]
    .groupby('neighborhood')
    .agg(grupo=('grupo', lambda x: x.value_counts().idxmax()))
    .reset_index()
)

series_hex_f = series_hex_f.merge(mapping_hex, on='hex_id')
series_colonia_f = series_colonia_f.merge(mapping_colonia, on='neighborhood')
```

```python
# 8. Validación de tendencias paralelas (solo Céntrico vs Periférico)
def grafica_tendencias(df, id_col, label, corte='2020-03-01'):
    df['post'] = (df['period'] >= corte).astype(int)
    dfg = df.groupby(['period', 'grupo'])['avg_price_premium'].mean().reset_index()
    plt.figure(figsize=(10,5))
    for g, l in zip([0,1], ['Control','Tratamiento']):
        plt.plot(dfg[dfg['grupo']==g]['period'], dfg[dfg['grupo']==g]['avg_price_premium'], label=l)
    plt.axvline(pd.Timestamp(corte), color='red', linestyle='--', label='Inicio COVID')
    plt.title(f'Tendencias paralelas: {label}')
    plt.legend()
    plt.show()
grafica_tendencias(series_hex_f, 'hex_id', 'Hexágono')
grafica_tendencias(series_colonia_f, 'neighborhood', 'Colonia')
```

```python
# --- 8. Validación de tendencias paralelas ---
def grafica_tendencias(df, id_col, label, corte='2020-03-01'):
    df['post'] = (df['period'] >= corte).astype(int)
    dfg = df.groupby(['period', 'grupo'])['avg_price_premium'].mean().reset_index()
    plt.figure(figsize=(10,5))
    for g, l in zip([0,1], ['Control','Tratamiento']):
        plt.plot(dfg[dfg['grupo']==g]['period'], dfg[dfg['grupo']==g]['avg_price_premium'], label=l)
    plt.axvline(pd.Timestamp(corte), color='red', linestyle='--', label='Inicio COVID')
    plt.title(f'Tendencias paralelas: {label}')
    plt.legend()
    plt.show()

grafica_tendencias(series_hex_f, 'hex_id', 'Hexágono')
grafica_tendencias(series_colonia_f, 'neighborhood', 'Colonia')
```

```python
# 7. Aplicación de DID en ambas escalas
def run_did(df, id_col, corte='2020-03-01'):
    df = df.copy()
    df['post'] = (df['period'] >= corte).astype(int)
    modelo = smf.ols('avg_price_premium ~ grupo * post', data=df).fit()
    efecto = modelo.params['grupo:post']
    pval = modelo.pvalues['grupo:post']
    return efecto, pval, modelo.conf_int().loc['grupo:post'].tolist()
efecto_hex, pval_hex, ci_hex = run_did(series_hex_f, 'hex_id')
efecto_col, pval_col, ci_col = run_did(series_colonia_f, 'neighborhood')

print(f"Efecto causal DID (hexágono): {efecto_hex:.4f}, p-valor: {pval_hex:.4f}, IC95%: {ci_hex}")
print(f"Efecto causal DID (colonia): {efecto_col:.4f}, p-valor: {pval_col:.4f}, IC95%: {ci_col}")
```

**¿Qué hacer con "Intermedio"?**
La literatura internacional recomienda:

No incluir "Intermedio" en el análisis principal si buscas una estimación clara y robusta del efecto causal promedio, especialmente cuando su definición es ambigua o representa una mezcla heterogénea de exposición al shock.

Esto ayuda a evitar contaminación entre grupos y asegura un contraste más limpio y comparable.

Usar "Intermedio" en análisis de sensibilidad o exploratorios:

Puedes incluirlo como grupo adicional en modelos DID con tres categorías, o como parte de un análisis de gradientes espaciales.

Esto permite investigar si existe un patrón de transición o gradiente entre centro y periferia, o si el efecto en "Intermedio" es más parecido al control o al tratamiento.

**¿Cómo proceder?**
Análisis principal:
Mantén la comparación entre "Céntrico" y "Periférico" para el reporte principal y la interpretación causal central.

**Análisis complementario:**
Incluye "Intermedio" en modelos extendidos (DID con tres grupos o multinivel) para explorar posibles gradientes o patrones espaciales adicionales. Reporta estos resultados como análisis de robustez o heterogeneidad.

#### ¿Deberías realizar un análisis de potencia para confirmar la falta de efectos significativos?
¿Por qué es relevante el análisis de potencia?
Cuando los resultados de un análisis causal (como DID) muestran efectos no significativos, existen dos posibles explicaciones:

El efecto real es nulo o muy pequeño.

El estudio no tiene suficiente potencia estadística (es decir, no tiene suficiente tamaño de muestra, variabilidad, o contraste entre grupos) para detectar un efecto que sí existe.

Un análisis de potencia (power analysis) permite distinguir entre estos escenarios, evaluando si el diseño del estudio tenía capacidad suficiente para detectar efectos de magnitud relevante según la literatura o la práctica.

¿Qué recomienda la literatura internacional?
Zeng & Yi (2022), Tomal & Helbich (2022) y Tsai et al. (2022) enfatizan la importancia de evaluar la potencia estadística en estudios de impacto de shocks exógenos, especialmente cuando los efectos estimados son pequeños o no significativos.

Un análisis de potencia es altamente recomendable para justificar que la “ausencia de evidencia” no se debe a limitaciones del diseño, sino a la verdadera ausencia de un efecto relevante.

¿Qué implica hacer un análisis de potencia en tu caso?
Determinar el tamaño mínimo de efecto detectable (Minimum Detectable Effect, MDE) dado tu tamaño de muestra, varianza y nivel de significancia.

Simular o calcular la potencia para efectos de magnitud relevante (por ejemplo, una diferencia del 1% o 2% en el premio de precio).

Si la potencia es baja, deberías reportar la limitación y evitar concluir que “no hay efecto”.

Si la potencia es alta, puedes concluir con mayor confianza que la falta de significancia refleja una ausencia de efecto relevante.

¿Cómo implementarlo en la práctica?
Calcula la potencia post-hoc para el modelo DID usando el tamaño de muestra, la varianza residual y el tamaño de efecto observado o relevante.

Herramientas recomendadas:

Paquetes como statsmodels.stats.power en Python (TTestIndPower para diferencias de medias, o simulaciones para modelos lineales).

Simulación Monte Carlo para modelos más complejos.

```python
from statsmodels.stats.power import TTestIndPower

# Parámetros ejemplo
effect_size = 0.0025  # Tamaño de efecto relevante (ajusta según tu contexto)
alpha = 0.05          # Nivel de significancia
nobs1 = df_cdmx[df_cdmx['grupo'] == 1].shape[0]  # Observaciones en tratamiento
nobs2 = df_cdmx[df_cdmx['grupo'] == 0].shape[0]  # Observaciones en control

# Subconjuntos
premio_trat = df_cdmx[df_cdmx['grupo'] == 1]['price_premium']
premio_ctrl = df_cdmx[df_cdmx['grupo'] == 0]['price_premium']

# Desviaciones estándar y tamaños de muestra
std1 = premio_trat.std(ddof=1)
std2 = premio_ctrl.std(ddof=1)
n1 = premio_trat.shape[0]
n2 = premio_ctrl.shape[0]

# Desviación estándar combinada (pooled)
pooled_std = np.sqrt(((n1 - 1)*std1**2 + (n2 - 1)*std2**2) / (n1 + n2 - 2))
std = pooled_std      # Desviación estándar combinada

# Calcula la potencia
analysis = TTestIndPower()
power = analysis.solve_power(effect_size=effect_size/std, nobs1=nobs1, alpha=alpha, ratio=nobs2/nobs1, alternative='two-sided')
print(f'Potencia estadística para detectar un efecto de {effect_size}: {power:.2f}')
```

### Análisis de Potencia Estadística en el DID
Resumen del resultado
Potencia estadística obtenida: 0.82

Tamaño de efecto mínimo relevante evaluado: 0.0025 (en la métrica de premio de precio ajustado)

Interpretación
Una potencia estadística de 0.82 indica que el estudio tiene un 82% de probabilidad de detectar un efecto real de al menos 0.0025 si este existe.

El umbral de referencia internacional para un estudio bien diseñado es una potencia de 0.80 o superior. Por encima de este valor, se considera que el análisis tiene suficiente capacidad para identificar efectos relevantes y minimizar el riesgo de error tipo II (falso negativo).

Implicaciones para la robustez de los resultados
La ausencia de efectos significativos en tu análisis DID no se debe a falta de potencia estadística, sino a que el efecto real, si existe, es menor al umbral considerado relevante o verdaderamente nulo.

Puedes concluir con confianza que no hay evidencia de un efecto causal relevante de la pandemia sobre el premio de precio en los grupos comparados, bajo el diseño y los datos actuales.

Recomendaciones
Reporta explícitamente la potencia alcanzada en tu estudio para fortalecer la credibilidad de tus conclusiones.

Si en futuras versiones deseas detectar efectos aún más pequeños, sería necesario incrementar el tamaño de muestra o reducir la varianza residual.

Respaldo en la literatura
Estudios recientes sobre el impacto de shocks en mercados inmobiliarios enfatizan la importancia de reportar y justificar la potencia estadística en análisis causales, especialmente cuando los resultados no son significativos.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import mixedlm

# 1. Carga y preparación ---------------------------------------------------
df = pd.read_parquet('../data/properties_predicted.parquet')
df['valuation_date'] = pd.to_datetime(df['valuation_date'])

# 2. Calcular price premium y periodo --------------------------------------
df['log_y']      = np.log(df['y'])
df['log_y_pred'] = np.log(df['y_pred'])
df['price_premium'] = df['log_y'] - df['log_y_pred']
df['period']     = df['valuation_date'].dt.to_period('M')

# 3. Filtrar CDMX ----------------------------------------------------------
df_cdmx = df[df['state_id'].astype(str) == '09'].copy()

# 4. Distancia al centro y zonas -------------------------------------------
centro = {'lon': -99.1332, 'lat': 19.4326}
def haversine_np(lon1, lat1, lon2, lat2):
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon, dlat = lon2-lon1, lat2-lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

df_cdmx['dist_centro_km'] = haversine_np(
    df_cdmx['longitude'], df_cdmx['latitude'],
    centro['lon'], centro['lat']
)
umbral_c, umbral_i = 6, 12
df_cdmx['zona']  = pd.cut(
    df_cdmx['dist_centro_km'],
    bins=[-np.inf, umbral_c, umbral_i, np.inf],
    labels=['Céntrico','Intermedio','Periférico']
)
df_cdmx['grupo'] = df_cdmx['zona'].map({'Céntrico':1,'Periférico':0})

# 5. Series de tiempo ------------------------------------------------------
def make_series(df, id_col):
    ser = (df
           .groupby([id_col,'period'])
           .agg(avg_price_premium=('price_premium','mean'),
                volume=('y','count'))
           .reset_index())
    ser['period'] = ser['period'].dt.to_timestamp()
    return ser

series_hex = make_series(df_cdmx, 'hex_id')
series_col = make_series(df_cdmx, 'neighborhood')

# 6. Filtrado de paneles ---------------------------------------------------
def filtra_panel(df, id_col, min_pre=6, min_post=6, corte='2020-03-01'):
    pre  = df[df['period']<corte].groupby(id_col)['volume'].sum()
    post = df[df['period']>=corte].groupby(id_col)['volume'].sum()
    ok   = pre[pre>=min_pre].index.intersection(post[post>=min_post].index)
    return df[df[id_col].isin(ok)].copy()

series_hex_f = filtra_panel(series_hex, 'hex_id')
series_col_f = filtra_panel(series_col, 'neighborhood')

# 7. Asignar grupo consistente ---------------------------------------------
df_valid = df_cdmx.dropna(subset=['grupo'])
def mapping(df, id_col):
    m = (df
         .groupby(id_col)
         .agg(
            zona=('zona', lambda s: s.mode().iat[0] if not s.mode().empty else pd.NA),
            grupo=('grupo',lambda s: s.mode().iat[0] if not s.mode().empty else pd.NA)
          )
         .reset_index())
    return m

map_hex = mapping(df_valid, 'hex_id')
map_col = mapping(df_valid, 'neighborhood')

series_hex_f = series_hex_f.merge(map_hex[['hex_id','grupo']], on='hex_id', how='inner')
series_col_f = series_col_f.merge(map_col[['neighborhood','grupo']], on='neighborhood', how='inner')

# 8. Definir post-COVID ----------------------------------------------------
cutoff = pd.Timestamp('2020-03-01')
series_hex_f['post'] = (series_hex_f['period']>=cutoff).astype(int)
series_col_f['post'] = (series_col_f['period']>=cutoff).astype(int)

# 9. Modelos multinivel ----------------------------------------------------
# 9a) Hexágonos: intercepto + pendiente aleatoria en post
model_hex = mixedlm(
    "avg_price_premium ~ grupo * post",
    data=series_hex_f,
    groups=series_hex_f["hex_id"],
    re_formula="~post"
)
res_hex = model_hex.fit(reml=True)
print(res_hex.summary())

# 9b) Colonias: solo intercepto aleatorio
model_col = mixedlm(
    "avg_price_premium ~ grupo * post",
    data=series_col_f,
    groups=series_col_f["neighborhood"]
)
res_col = model_col.fit(reml=True)
print(res_col.summary())

# 10. Efectos aleatorios (hexágonos) ---------------------------------------
# Inspecciona nombres de efectos aleatorios para un grupo de ejemplo
example_gid = next(iter(res_hex.random_effects))
print("Random effect names:", res_hex.random_effects[example_gid].index.tolist())

# Monta DataFrame y renombra
df_re = pd.DataFrame(res_hex.random_effects).T
# el índice df_re.columns coincide con esos nombres: p.ej. ['Intercept','post']
# los renombramos:
df_re = df_re.rename(columns={df_re.columns[0]:'Intercept',
                              df_re.columns[1]:'Slope_post'})

print(df_re.head())

# 11. Histograma de pendientes post aleatorias -----------------------------
plt.figure(figsize=(6,4))
df_re['Slope_post'].hist(bins=30)
plt.title("Distribución de pendientes aleatorias en 'post' (hexágonos)")
plt.xlabel("Pendiente post")
plt.ylabel("Número de hexágonos")
plt.tight_layout()
plt.show()

# 12. Observado vs Ajuste multilevel (ejemplo) -----------------------------
hex_ej = series_hex_f['hex_id'].iloc[0]
sub    = series_hex_f[series_hex_f['hex_id']==hex_ej]

# Extrae intercepto y slope para este hexágono
re_series = res_hex.random_effects[hex_ej]
int_re   = re_series.iloc[0]
slope_re = re_series.iloc[1] if len(re_series)>1 else 0.0

# Construye la serie ajustada (fijos + aleatorios)
fe = res_hex.fe_params
fitted = (
     fe["Intercept"]
   + fe["grupo"]*sub["grupo"]
   + fe["post"]*sub["post"]
   + fe["grupo:post"]*sub["grupo"]*sub["post"]
   + int_re
   + slope_re*sub["post"]
)

plt.figure(figsize=(8,4))
plt.plot(sub["period"], sub["avg_price_premium"], label="Observado")
plt.plot(sub["period"], fitted,            label="Ajuste multilevel")
plt.axvline(cutoff, color="red", linestyle="--", label="COVID Inicio")
plt.legend()
plt.title(f"Hexágono {hex_ej}: Observado vs Ajuste Multilevel")
plt.tight_layout()
plt.show()
```

En estos dos gráficos podemos ver:

Distribución de pendientes aleatorias en “post” (hexágonos)

La mayoría de los hexágonos tiene una pendiente posterior al corte (marcada en marzo 2020) próxima a cero.

La distribución es casi simétrica alrededor de 0, con unas pocas áreas que experimentan subidas o bajadas pronunciadas, pero la mediana y la media de esas pendientes es prácticamente cero.

Ejemplo de un hexágono (89499580b6bffff): Observado vs Ajuste Multilevel

Tras el inicio de la pandemia (línea roja), el precio‐premium observado en ese hexágono oscila, pero el modelo multinivel predice una pendiente plana (línea naranja casi horizontal).

Es decir, el componente aleatorio “post” para ese hexágono (su desviación individual) se suma a un efecto fijo prácticamente nulo, y en conjunto no se ve un cambio sistemático atribuible a COVID en la prima de precios.

Conclusión general

El modelo de efectos mixtos (DID multinivel) nos da un coeficiente grupo × post muy cercano a cero y no estadísticamente significativo (p≫0.05).

En promedio, no encontramos un impacto causal claro del COVID-19 sobre la diferencia de log-precios (price premium) entre las zonas céntricas (tratamiento) y periféricas (control).

Existe heterogeneidad local —algunos hexágonos subieron más, otros bajaron más— pero la tendencia central es nula.

⚠️ Esto sugiere que, aunque hubo fluctuaciones puntuales en distintos micromercados, no hubo un efecto uniforme de la pandemia sobre la prima de precio en Ciudad de México, al menos a la escala hexagonal y controlando por la heterogeneidad espacial.
